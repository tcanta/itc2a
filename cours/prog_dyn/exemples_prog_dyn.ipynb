## Programmation Dynamique

### Factorielle

```python
def fact(n):
    if n == 0: # cas de base
        return 1
    return n*fact(n - 1) # appel récursif
```

```python
fact(42)
```

### Fibonacci

Version naive recursive

```python
def fibo(n):
    if n == 0 or n == 1:
        return 1
    return fibo(n - 1) + fibo(n - 2)
```

```python
fibo(32)
```

Version dynamique

```python
def fibo(n):
    L = [1, 1] # L[i] va contenir le ième terme de Fibonacci
    for i in range(n - 1):
        L.append(L[-1] + L[-2]) # récurrence
    return L[n]

def fibo(n):
    f0, f1 = 1, 1 # f0, f1 sont les deux derniers termes
    for i in range(n - 1):
        f0, f1 = f1, f0 + f1
    return f1
```

```python
fibo(32)
```

### Coefficients binomiaux

Version naive récursive

```python
def binom(n, k):
    if k == 0: return 1
    if n == 0: return 0
    return binom(n - 1, k - 1) + binom(n - 1, k)
```

```python
binom(30,8)
```

Version dynamique

```python
def binom(n, k):
    M = [[0]*(k + 1) for _ in range(n + 1)]
    for i in range(0, n + 1):
        M[i][0] = 1
    for i in range(1, n + 1):
        for j in range(1, k + 1):
            M[i][j] = M[i - 1][j - 1] + M[i - 1][j]
    return M[n][k]
```

```python
binom(30,8)
```

### Knapsack

```python
def knapsack(c, w, v):
    """
    Renvoie la valeur maximum que l'on peut mettre
    dans un sac à dos de capacité c.
    Le ième objet a pour poids w[i] et valeur v[i].
    """
    n = len(w) # nombre d'objets
    dp = [[0]*(n + 1) for i in range(c + 1)]
    # dp[i][j] = valeur max dans un sac de capacité i
    # où j est le nombre d'objets autorisés
    for i in range(1, c + 1):
        for j in range(1, n + 1):
            if w[j - 1] <= i:
                x = v[j - 1] + dp[i - w[j - 1]][j - 1]
                dp[i][j] = max(dp[i][j - 1], x)
            else:
                dp[i][j] = dp[i][j - 1]
    return dp[c][n]
```

```python
knapsack(7,[1,2,2,4,5],[1,4,5,2,4])
```

```python
def knapsack2(c, w, v):
    n = len(w)
    dp = [0]*(c + 1)
    for j in range(n):
        dp_ = dp[:] # copie de dp
        for i in range(c + 1):
            if w[j] <= i:
                dp[i] = max(dp_[i], v[j] + dp_[i - w[j]])
    return dp[-1]
```

```python
knapsack2(7,[1,2,2,4,5],[1,4,5,2,4])
```

## Mémoïsation

### Fibonacci

```python
def fibo(n):
    d = {} # d[k] contiendra le kème terme de la suite
    def aux(k):
        if k == 0 or k == 1:
            return 1
        if k not in d:
            d[k] = aux(k - 1) + aux(k - 2)
        return d[k]
    return aux(n)
```

```python
fibo(32)
```

### Coefficients Binomiaux

```python
def binom(n, k):
    d = {}
    def aux(i, j):
        if j == 0: return 1
        if i == 0: return 0
        if (i, j) not in d:
            d[(i, j)] = aux(i - 1, j - 1) + aux(i - 1, j)
        return d[(i, j)]
    return aux(n, k)
```

```python
binom(30,8)
```

### Knapsack

```python
def knapsack_memo(c, w, v):
    dp = {}
    def aux(i, j):
        if i == 0 or j == 0:
            return 0
        if (i, j) not in dp:
            dp[(i, j)] = aux(i, j - 1)
            if w[j - 1] <= i:
                x = v[j - 1] + aux(i - w[j - 1], j - 1)
                dp[(i, j)] = max(dp[(i, j)], x)
        return dp[(i, j)]
    return aux(c, len(w))
```

```python
knapsack_memo(7,[1,2,2,4,5],[1,4,5,2,4])
```
