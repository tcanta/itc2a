```python
def inverse(d):
    d2 = {}
    for k in d:
        d2[d[k]]=k
    return d2
```

```python
d = {0:1,1:4,7:6}
print(inverse(d))
```

```python
def frequent(L):
    d = {0:0}
    maxkey = 0
    for x in L:
        if x in d:
            d[x]+=1
            if d[x] > d[maxkey]:
                maxkey = x
        else:
            d[x]=1
            if d[maxkey]==0:
                maxkey=x
    return (maxkey,d[maxkey])
```

```python
l=[1,1,1,1,2,5,8,9,8,8,8,9,5,6,4,7,1,1]
frequent(l)
```

```python
def eratosthene(n):   
    s = {i for i in range(n+1)}
    c=2
    to_remove={0,1} # 0 n'est pas premier et 1 non plus !
    while c < n**0.5:
        for x in s:
            if x%c==0 and x!=c :
                to_remove.add(x)
        c+=1
    s=s.difference(to_remove)
    return s

eratosthene(100000)
```
